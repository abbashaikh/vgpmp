```python
import numpy as np
import tensorflow as tf

# start and end time
t0, tN  = 0.0, 1.0
# number of interior support points (K)
num_interior_points = 9
# times
t = np.linspace(t0, tN, num_interior_points + 2)
X_data = t.reshape(-1, 1).astype(np.float64)         # shape (N,1)

num_query_points = 20
X_test = np.linspace(t0, tN, num_query_points, dtype=np.float64).reshape(-1,1)

Xq = tf.convert_to_tensor(X_test)
Xd = tf.convert_to_tensor(X_data)

Xq = tf.reshape(Xq, (-1,))  # (M,)
Xd = tf.reshape(Xd, (-1,))  # (N,)

N = tf.shape(Xd)[0]

idx_next = tf.searchsorted(Xd, Xq, side="right")          # (M,)
i = tf.clip_by_value(idx_next - 1, 0, N - 2)              # (M,)

tf.print(i)
```
